```python
# Define function to extract name from columns 'keywords','genres',
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
# function to extract top 3 cast
def extract3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter <3:
            L.append(i['name'])
        counter +=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(extract3)
```

```python
# To fetch director name from crew
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job']=='Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] =  movies['crew'].apply(fetch_director)
```

```python
# To remove extra spaces
def remove_spaces(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1

# Apply the remove_spaces function to multiple columns
movies['cast'] = movies['cast'].apply(remove_spaces)
movies['crew'] = movies['crew'].apply(remove_spaces)
movies['genres'] = movies['genres'].apply(remove_spaces)
movies['keywords'] = movies['keywords'].apply(remove_spaces)
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
def remove_stopwords(obj):
    final_sentence = []
    tokens = word_tokenize(obj)

    for each_word in tokens:
        if each_word not in stopwords.words('english'):
            final_sentence.append(each_word)
    return " ".join(final_sentence)
```

```python
# Stemming
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stemming(text):
    result =[]
    for i in text.split():
        result.append(ps.stem(i))
    return " ".join(result)
```

```python
movies['tags'] =  movies['tags'].apply(stemming)
```

```python
def find_closest(text):
    spliter = movies['title'].str.split(' ')
    i = 0
    for val in spliter:
        if text in val:
            print(movies['title'][i])
            return i
        i+=1
```

```python
def recommend_with_bert(movie):
    movie = movie.lower()
    titles = movies['title']
    if movie in titles.to_list():
        index = movies[movies['title']==movie].index[0]
    else:
        index = find_closest(movie)

    distances = sorted(list(enumerate(similarity[index])), reverse=True, key = lambda x: x[1])
    for i in distances[1:10]:
        print(movies.iloc[i[0]].title)
```
